```python
filename_100kOhm = "Band-pass_filter.csv"
filename_10kOhm = "Band-pass_filter_10k.csv"
```

```python
data = np.random.rand(100)
plt.plot(data)
```

```python
data = np.loadtxt(filename_100kOhm,delimiter = ',',skiprows = 22,usecols= (0,1,2,3))
plt.figure("attempt 1")
plt.plot(data[:,0]/1000,data[:,2],label="100k $\Omega$",color = "black")
plt.semilogx()
plt.xlabel("Frequency(kHz)")
plt.ylabel("Decibels(dB)")
plt.grid()
plt.legend()


ind = np.argmax(data[:,2])
f0 = data[ind,0]

plt.xlim(48,50)
plt.ylim(-24,-17)
plt.vlines(f0/1000, -24,-15,color = "red")
```

```python
data = np.loadtxt(filename_10kOhm,delimiter = ',',skiprows = 22,usecols= (0,1,2,3))
plt.figure("attempt 2")
plt.plot(data[:,0]/1000,data[:,2],label = "10k$\Omega$")
plt.semilogx()
plt.xlabel("Frequency(kHz)")
plt.ylabel("Decibels(dB)")
plt.grid()
plt.legend()

ind = np.argmax(data[:,2])
f0 = data[ind,0]

plt.xlim(47.5,50.3)
plt.ylim(-8,-4.5)
plt.vlines(f0/1000, -4,-8,color = "red")
```
